```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("smart-panda/food.csv")
df["day"] = df["time"].apply(lambda time: time.split(" ")[0])
df2 = df[df.type == "milk"].groupby(by=["day"])["size"].agg({"size": np.sum}).reindex()

df2[:].plot.bar()
```
